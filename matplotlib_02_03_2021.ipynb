```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Настройка стилей
plt.style.use("seaborn")

x = np.linspace(1, 10)
y = 1 / x

# создаём фигуру
figure = plt.figure()

# добавляем график
plt.plot(y, '--', color="red")
```

```python
import os
import os.path 

if not os.path.exists("new_plot"):
    figure.savefig("new_plot.png")

print(os.listdir())
```
